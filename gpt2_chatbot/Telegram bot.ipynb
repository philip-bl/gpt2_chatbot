```
import random

dialogs = {}
status = {}
votes = []
votes_tmp = {}

    
with open("gdrive/My Drive/dialogs","r",encoding="utf-8") as file:
    dialogs = json.load(file)
    
    
with open("gdrive/My Drive/users","r",encoding="utf-8") as file:
      users = json.load(file)
    
with open("gdrive/My Drive/status","r",encoding="utf-8") as file:
      status = json.load(file)
    
    
    
  
import time
while True:
    try:
        with open("gdrive/My Drive/dialogs","r",encoding="utf-8") as file:
            dialogs = json.load(file)


        with open("gdrive/My Drive/users","r",encoding="utf-8") as file:
              users = json.load(file)

        with open("gdrive/My Drive/status","r",encoding="utf-8") as file:
              status = json.load(file)
            
            
        bot = telebot.TeleBot('839509606:AAHUWMhI0bimh-OU8heyn87zBfVspakvBGk')
        
        
        @bot.message_handler(content_types=['text'])
        def get_text_messages(message):
          
            id = message.from_user.id

            users[str(id)] = message.from_user.username
            print(users)
            with open("gdrive/My Drive/users","w",encoding="utf-8") as file:
              json.dump(users,file)
              
            with open("gdrive/My Drive/dialogs","r",encoding="utf-8") as file:
              dialogs = json.load(file)
              
            with open("gdrive/My Drive/status","r",encoding="utf-8") as file:
              status = json.load(file)

            #if id == 144346271:
            if message.text == "/start":
              status[str(id)] = random.choice([0,1,2])
              with open("gdrive/My Drive/status","w",encoding="utf-8") as file:
                json.dump(status,file)
              bot.send_message(id, "Just chat with neural bot. After a short conversation you will have a chance to evaluate the quality of our AI model.\
                               Sometimes after reseting the neural model changes (we have several), so try all of them! \n/reset restarts the conversation\n(don't hesitate to reset it to delete the previous context from the bot memory).")

            elif message.text == "/reset":
              dialogs[str(id)] = []
              status[str(id)] = random.choice([0,1,2])
              
              with open("gdrive/My Drive/status","w",encoding="utf-8") as file:
                json.dump(status,file)
                
              with open("gdrive/My Drive/dialogs","w",encoding="utf-8") as file:
                  json.dump(dialogs,file)
                  
              bot.send_message(id,  "Just chat with neural bot. After a short conversation you will have a chance to evaluate the quality of our AI model.\
                               Sometimes after reseting the neural model changes (we have several), so try all of them!\n/reset restarts the conversation\n(don't hesitate to reset it to delete the previous context from the bot memory)")

            elif message.text == "/help":
              bot.send_message(id, "/dialogs prints all messages, /reset restarts the conversation")

            elif message.text == "/dialogs":
              for i in dialogs[str(id)]:
                bot.send_message(id, i)

            elif message.text == "/set1":
              bot.send_message(id, "now original GPT2 is active")
              status[str(id)] = 0
              with open("gdrive/My Drive/status","w",encoding="utf-8") as file:
                json.dump(status,file)
              dialogs[str(id)] = []
              with open("gdrive/My Drive/dialogs","w",encoding="utf-8") as file:
                  json.dump(dialogs,file)
                  
            elif message.text == "/set2":
              bot.send_message(id, "now  new underfited  GPT2 is active")
              status[str(id)] = 1
              with open("gdrive/My Drive/status","w",encoding="utf-8") as file:
                json.dump(status,file)
              dialogs[str(id)] = []
              with open("gdrive/My Drive/dialogs","w",encoding="utf-8") as file:
                  json.dump(dialogs,file)
                  
            elif message.text == "/set3":
              bot.send_message(id, "now Philipp's GPT2 is on")
              status[str(id)] = 2
              with open("gdrive/My Drive/status","w",encoding="utf-8") as file:
                json.dump(status,file)
              dialogs[str(id)] = []
              with open("gdrive/My Drive/dialogs","w",encoding="utf-8") as file:
                  json.dump(dialogs,file)
                  
            elif message.text == "/set4":
              bot.send_message(id, "now underfited GPT2 is on")
              status[str(id)] = 3
              with open("gdrive/My Drive/status","w",encoding="utf-8") as file:
                json.dump(status,file)
              dialogs[str(id)] = []
              with open("gdrive/My Drive/dialogs","w",encoding="utf-8") as file:
                  json.dump(dialogs,file)
                  

            else:
              with open("gdrive/My Drive/dialogs","r",encoding="utf-8") as file:
                dialogs = json.load(file)
              dialog = dialogs.get(str(id),[])
              
              with open("gdrive/My Drive/status","r",encoding="utf-8") as file:
                status = json.load(file)
                
              state = int(status.get(str(id),2))
              model = models[state]
              
              try:
                response = produce_answer(message.text, dialog,50,2, 10,1, False, model, enc, device,insert_intro=True, wrap_type='name')#"Hi, I am Bob, the bot."
                with open("gdrive/My Drive/dialogs","w",encoding="utf-8") as file:
                  json.dump(dialogs,file)
                
              except:
                print("AAAAAAAAAAAAAAAAAAAAAAA")
                dialogs[str(id)] = []
                response = produce_answer("...", dialog,50,2,10,1,False,model, enc, device,insert_intro=True, wrap_type='name')#"Hi, I am Bob, the bot."
                
                with open("gdrive/My Drive/dialogs","w",encoding="utf-8") as file:
                  json.dump(dialogs,file)
     
              bot.send_message(id, response)


            with open("gdrive/My Drive/dialogs","r",encoding="utf-8") as file:
                dialogs = json.load(file)
            
            if len(dialogs.get(str(id),[])) > 14:
              vote = telebot.types.ReplyKeyboardMarkup()
              vote.row("1","2","3","4","5")
              bot.send_message(message.chat.id, "PLEASE EVELUATE THE QUALITY OF THE CONVERSATION:\n How clever is the bot?\n0 - stupid bot \n5 - bot is smarter then stupid humans", reply_markup=vote)
              bot.register_next_step_handler(message, poll_2);
              
              with open("gdrive/My Drive/dialogs","w",encoding="utf-8") as file:
                json.dump(dialogs,file)






        def poll_2(message):
              id = message.from_user.id
              vote = telebot.types.ReplyKeyboardMarkup()
              vote.row("1","2","3","4","5")
              bot.send_message(message.chat.id, " Did the bot understood context of your conversation correctly?\n0 - stupid bot \n5 - bot is smarter then stupid humans", reply_markup=vote)
              bot.register_next_step_handler(message, poll_3);

              with open("gdrive/My Drive/votes_tmp","r",encoding="utf-8") as file:
                votes_tmp = json.load(file)
                
              votes_tmp[str(id)] =[]
              votes_tmp[str(id)].append(message.text)

              with open("gdrive/My Drive/votes_tmp","w",encoding="utf-8") as file:
                json.dump(votes_tmp,file)

        def poll_3(message):
              id = message.from_user.id
              vote = telebot.types.ReplyKeyboardMarkup()
              vote.row("1","2","3","4","5")
              bot.send_message(message.chat.id, " How human-like was the bot?\n0 - stupid bot \n5 - bot is more human then stupid humans", reply_markup=vote)
              bot.register_next_step_handler(message, poll_end);

              with open("gdrive/My Drive/votes_tmp","r",encoding="utf-8") as file:
                votes_tmp = json.load(file)
              votes_tmp[str(id)].append(message.text)

              with open("gdrive/My Drive/votes_tmp","w",encoding="utf-8") as file:
                json.dump(votes_tmp,file)      


        def poll_end(message):
          with open("gdrive/My Drive/dialogs","r",encoding="utf-8") as file:
            dialogs = json.load(file)
            
          with open("gdrive/My Drive/status","r",encoding="utf-8") as file:
              status = json.load(file)
          #global dialogs
          #global status
          id = message.chat.id

          with open("gdrive/My Drive/votes","r",encoding="utf-8") as file:
            votes = json.load(file)
            
          state = status.get(str(message.chat.id),"hz")


          with open("gdrive/My Drive/votes_tmp","r",encoding="utf-8") as file:
            votes_tmp = json.load(file)
            
          votes_tmp[str(id)].append(message.text)

          votes.append([state,votes_tmp[str(id)],message.chat.id,dialogs[str(message.chat.id)]])

          with open("gdrive/My Drive/votes","w",encoding="utf-8") as file:
            json.dump(votes,file)
            
          dialogs[str(message.chat.id)] = []
          status[str(message.chat.id)] = random.choice([0,1,2])
          
          with open("gdrive/My Drive/dialogs","w",encoding="utf-8") as file:
            json.dump(dialogs,file)
            
          with open("gdrive/My Drive/status","w",encoding="utf-8") as file:
            json.dump(status,file)

          bot.send_message(message.chat.id, "Thank you! Your conversation has been reseted.", reply_markup=telebot.types.ReplyKeyboardRemove())
  
 


        
        bot.polling(none_stop=True)
    except Exception:
        time.sleep(3)
```
